# filling the null values

```python
df2= df.fillna(value=0)
df2
```

```python
df2.isnull().sum().sum()
```

# filling null values with the previous value

```python
df4= df.ffill()                  #df.fillna(method= 'pad')
df4
```

# null values fill with their next value

```python
df5= df.bfill()
df5                                         # df.fillna(method= 'bfill')
```

```python
df6=df.fillna(method= 'bfill', axis=1) # column wise fill the null values 
df6
```

# filling different values in null in different columns

```python
df7= df.fillna({'society': 'abcd',
                'balcony': 'efgh'})
df7
```

# filling null values with the mean of column

```python
df8= df.fillna(value= df['balcony'].mean())
df8
```

```python
df9= df.fillna(value= df['balcony'].max()) # with max value
df9
```

```python
df10= df.fillna(value= df['balcony'].min()) # with thier min  value
df10
```

# dropna working 

```python
df11= df.dropna()
df11
```

```python
df12= df.dropna(how= 'any')   # if any null values are present in one column/row, then it will remove them
df12
```

```python
df13= df.dropna(how= 'all')  # if all null values present in one column/row, then it will remove them 
df13
```
